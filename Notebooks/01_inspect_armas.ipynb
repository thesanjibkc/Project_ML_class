```python
armas1 = armas1.rename(columns={'Index_D': 'Index_K'})
armas2 = armas2.rename(columns={'Index_D': 'Index_K'})
armas3 = armas3.rename(columns={'Index_D': 'Index_K'})
armas_all = armas_all.rename(columns={'Index_D': 'Index_K'})
```

```python
# STEP 1: Rename Index_D → Index_K (in-place)
for df in [armas1, armas2, armas3, armas_all]:
    df.rename(columns={'Index_D': 'Index_K'}, inplace=True)

# STEP 2: Create subplot grid (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# STEP 3: Define colors for bins
colors = ['#4CAF50', '#2196F3', '#FF9800']

# STEP 4: Store datasets with names
dfs = [
    ('armas1', armas1),
    ('armas2', armas2),
    ('armas3', armas3),
    ('armas_all', armas_all)
]

# STEP 5: Flatten axes (important)
axes = axes.flatten()

# STEP 6: Loop and plot
for i, (name, df) in enumerate(dfs):
    
    # Count values
    counts = df['Index_K'].value_counts().sort_index()
    
    # Plot bar chart
    counts.plot(
        kind='bar',
        ax=axes[i],
        color=colors
    )
    
    # Special styling for armas_all
    if name == 'armas_all':
        axes[i].set_title(name.upper(), fontsize=12, fontweight='bold')
        axes[i].set_facecolor('#f0f0f0')
    else:
        axes[i].set_title(name)
    
    # Labels
    axes[i].set_xlabel('Index_K')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)

    # STEP 7: Add values on top of bars
    for j, value in enumerate(counts):
        axes[i].text(
            j, value,
            str(value),
            ha='center',
            va='bottom'
        )

# STEP 8: Adjust layout
plt.tight_layout()

# plt.savefig(r'D:\Project_ML_class\Results\charts\Index_K_distribution.png', dpi=300)

# STEP 9: Show plot
plt.show()
```

```python
import numpy as np

def add_nairas_index(df, col_name):
    conditions = [
        (df[col_name] >= 0) & (df[col_name] < 5),
        (df[col_name] >= 5) & (df[col_name] < 15),
        (df[col_name] >= 15) & (df[col_name] <= 45)
    ]
    
    df[f'Index_K_{col_name}'] = np.select(conditions, [0, 1, 2], default=np.nan)
    
    return df
   
```

```python
for df in [armas1, armas2, armas3, armas_all]:
    add_nairas_index(df, 'NAIRASV3')
    add_nairas_index(df, 'NAIRASV3')
```

```python


# STEP 1: Create subplot grid (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# STEP 2: Define colors for bins
colors = ['#4CAF50', '#2196F3', '#FF9800']

# STEP 3: Store datasets with names
dfs = [
    ('armas1', armas1),
    ('armas2', armas2),
    ('armas3', armas3),
    ('armas_all', armas_all)
]

# STEP 4: Flatten axes
axes = axes.flatten()

# STEP 5: Loop and plot
for i, (name, df) in enumerate(dfs):
    
    # Count values from NAIRAS-based index
    counts = df['Index_K_NAIRASV3'].value_counts().sort_index()
    
    # Plot bar chart
    counts.plot(
        kind='bar',
        ax=axes[i],
        color=colors
    )
    
    # Special styling for armas_all
    if name == 'armas_all':
        axes[i].set_title(name.upper() + ' - NAIRASV3', fontsize=12, fontweight='bold')
        axes[i].set_facecolor('#f0f0f0')
    else:
        axes[i].set_title(name + ' - NAIRASV3')
    
    # Labels
    axes[i].set_xlabel('Index_K from NAIRASV3')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)

    # Add values on top of bars
    for j, value in enumerate(counts):
        axes[i].text(
            j, value,
            str(value),
            ha='center',
            va='bottom'
        )

# STEP 6: Adjust layout
plt.tight_layout()

# STEP 7: Show plot
plt.show()
```
